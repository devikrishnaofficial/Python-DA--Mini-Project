```python
import pandas as pd
```

```python
data=pd.read_csv("C:\\Users\\devik\\Downloads\\sales\\supermarket_sales - Sheet1.csv")
data.head(5)
```

```python
#Data Cleaning
print(data.isnull().sum())
data=data.dropna()
print(data.duplicated().sum())
data=data.drop_duplicates()
```

```python
#Data Transformation
data = data.sort_values(by='Date')
data.head(5)
```

```python
health_beauty = data[data['Product line'] == 'Health and beauty']
health_beauty
```

```python
Normal_Customer=data[data['Customer type']=="Normal"]
Normal_Customer
```

```python
#Aggregation
# Average sales per product line
average_sales_per_product = data.groupby('Product line')['Total'].mean()

# Sales counts per region
sales_counts_per_region = data.groupby('Branch')['Total'].count()


print("\nAverage Sales Per Product Line:")
print(average_sales_per_product)

print("\nSales Counts Per Region:")
print(sales_counts_per_region)
```
